# API: Application Program Interface
- ¿Que son?: mecanismo que permiten transferir informacion de forma eficiente y escalable
- ¿Para que sirven? aislan procesos, hacen que los flujos de trabajo sean modularizados, generan seguridad
- ¿Quien las crea?: desarrolladores, empresas o grupos 

**Clasificacion general**

- Publicas
- Privadas (necsitan ya API Key, API Secret o mecanismo OAuth)

**Clasificacion especifica**
- Desarrollador paga 
- Desarrollador gana

**Rest APIS (RestFul APIS)> REST: Representational State Transfer.** 
- **GET**: extraer datos sin enviar nada 
- **POST**: se envian parametros y devuelve resultados
- **PUT**: actualizaciones sobre datos
- **DELETE**: borrar datos

- FastAPI
- Flask
- Django
