# 2. EDA (Exploratory Data Analysis)

```python
C1=T[['Sex','Salary','RaceDesc']]
sns.boxplot(data=C1, x= 'Sex', y='Salary',showfliers= False)
plt.xlabel('Genero')
plt.ylabel('Salario')
plt.title('Boxplot de genero')
```

```python
C1=T[['Sex','Salary','RaceDesc']]
sns.boxplot(data=C1, x= 'Sex', y='Salary',hue= 'RaceDesc',showfliers= False)
plt.xlabel('Genero')
plt.ylabel('Salario')
plt.title('Boxplot de genero')
```

- El rango de los hombres es mayor que el de las mujeres
- Ligeramente mayor mediana de los hombres 
- Aunque hay muchos mas atipicos en las mujeres

```python
sns.boxplot(data=C1, x= 'Sex', y='Salary',showfliers= True)
```
