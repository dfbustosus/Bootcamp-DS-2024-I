# 3.1 Relplot

```python
sns.relplot(data=df, x='Absences',y='Salary',kind='scatter')
```

```python
sns.scatterplot(data=df, x='Absences',y='Salary')
```

```python
df['DateofHire']=pd.to_datetime(df['DateofHire'])
df['FechaSimple']=df['DateofHire'].dt.strftime('%Y-%m')
tabla_serie= df[['FechaSimple','Salary']].groupby(by=['FechaSimple']).count().reset_index()
plt.figure(figsize=(18,6))
#fig, ax= plt.subplots(figsize=(18,6))
sns.relplot(data=tabla_serie, x='FechaSimple',y='Salary',kind='line', )
plt.xticks(rotation=90)
plt.title('Serie de tiempo')
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=tabla_serie, x='FechaSimple',y='Salary')
plt.xticks(rotation=90)
plt.title('Serie de tiempo')
```
